```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set_theme(style = "white")
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
# Coletando Banco
uri = 'https://raw.githubusercontent.com/GeovaneBarros/Projeto-Data-Science/master/dataset/autos.csv'
venda_carros = pd.read_csv(uri)
```

```python
#Criacao plot
fig, ax = plt.subplots(figsize=(8,6))
sns.distplot(venda_carros['yearOfRegistration'],)
ax.set_title('Distribuição de veículos ao longo dos anos', fontsize= 15)
plt.xlabel("Ano de Registro")
plt.show()
```

```python
#Agrupando carros por tipos e ordenando
gruped = venda_carros.groupby('vehicleType')
gruped = gruped.size()
gruped = gruped.sort_values()

#Criando plot
sns.set_theme(style='darkgrid')
ax = sns.countplot(x='vehicleType', data=venda_carros, palette = 'GnBu_d', order = gruped.index)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45, horizontalalignment = 'right')
ax.set_title('Classificação de venda por tipo de veículo', fontsize = 15)
ax.set_ylabel('Quantidade de carros')
ax.set_xlabel('Tipos de carros')
plt.show()
```

```python
gruped = venda_carros.groupby('brand')
gruped = gruped.size()
gruped = gruped.sort_values()

fig, ax = plt.subplots(figsize=(8,10))
sns.countplot(y='brand', data=venda_carros, palette ='Blues', order = gruped.index)
ax.set_title('Classificação por marca', fontsize = 15)
ax.set_ylabel('Marca')
ax.set_xlabel('Quantidade')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax = sns.barplot( data = venda_carros, x='fuelType', y='price', hue = 'gearbox', ci=None)
ax.set_title('Preços de carros separado por combustível')
ax.set_ylabel('Preço')
ax.set_xlabel('Tipo de combustível')
plt.show()
```
