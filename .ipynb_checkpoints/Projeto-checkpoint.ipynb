```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
uri = 'https://raw.githubusercontent.com/GeovaneBarros/Projeto-Data-Science/master/dataset/autos.csv'
venda_carros = pd.read_csv(uri)
```

```python
anos_registro = venda_carros['yearOfRegistration']


anos_unico = anos_registro.drop_duplicates()


tipos_carros = venda_carros['vehicleType']
```

```python
fig, ax = plt.subplots(figsize=(8,6))
sns.distplot(anos_registro)
ax.set_title('Distribuição de veículos ao longo dos anos', fontsize= 15)
plt.xlabel("Ano de Registro")
plt.show()
```

```python
sns.set_theme(style='darkgrid')
ax = sns.countplot(x='vehicleType', data=venda_carros, palette = 'GnBu_d')
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45, horizontalalignment = 'right')
ax.set_title('Classificação de venda por tipo de veículo', fontsize = 15)
ax.set_ylabel('Quantidade de carros')
ax.set_xlabel('Tipos de carros')
plt.show()
```
